```python
def is_leap(year):
    """ return true for leap years, False for non leap years """
    return year % 4 == 0 and ( year % 100 != 0 or year % 400 == 0)


def chunks(lst, n):
    """Yield successive n-sized chunks from lst."""
    for i in range(0, len(lst), n):
        yield lst[i:i + n]
```

## Random Generation of Precipitation Data:

```python
def random_pcp(df,FirstYear, LastYear, ScenarioNumbers, RCPNames):
    
    #outFileName = OutFileName
    dfCol = df.columns
    #sceNum = len(dfCol)

    sceNum = ScenarioNumbers
    firstYear = FirstYear
    lastYear = LastYear
    simLen = lastYear - firstYear + 1

    from random import shuffle
    a = []
    for x in range(simLen): 
        randomInd = [z for z in range(sceNum)]
        shuffle(randomInd)
        a.append(randomInd)
        
    RCP = RCPNames
    columnsDf = ['sc_' + RCP + str(k) for k in range(1, sceNum+1,1)] 
    c = [[19810101 for p in range(sceNum)]]
    #df1 = 'df' + str(outDFNumber)
    df1 = pd.DataFrame(c, columns =columnsDf)
    #df1.to_csv('SAeidVaghefimodified1111222.csv', index = False)

    c.clear()

    i = 0
    p = 1
    for m in range(firstYear, lastYear + 1, 1):
        if is_leap(m):
            for j in range(((m - firstYear)*365+p) , ((m - firstYear)*365+367+p-1), 1):
                c.append(df[dfCol[a[i]]].iloc[j].values) 
            i += 1 # counter i; equal to simulation length (simLen)
            p += 1

        else:
            for j in range(((m - firstYear)*365+p), ((m - firstYear)*365+366+p-1), 1):
                c.append(df[dfCol[a[i]]].iloc[j].values) 
            i += 1
                
        print(m)

        dfnew = 'df' + str(m)
        dfnew = pd.DataFrame(c, columns =columnsDf)
        c.clear()
        df1 = df1.append(dfnew, ignore_index=True)
        #df1.to_csv('SAmodified1111222.csv', mode='w', header=True, index = False)
    return df1
```

Both maximum and minimum temperature data are being shuffeled. Resampling is done for each year independently.
